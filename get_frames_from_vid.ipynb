```python
import cv2
import os
from google.colab import files
from google.colab import drive
from PIL import Image
import numpy as np
from google.colab.patches import cv2_imshow
```

```python
path_to_save = 'spoof_print'
if not os.path.isdir(path_to_save):
        os.mkdir(path_to_save)

cap = cv2.VideoCapture('/content/gdrive/MyDrive/WIN_20220904_16_38_01_Pro.mp4')

fps = cap.get(cv2.CAP_PROP_FPS)
print('default fps:', fps)

count=0
while True:
    is_read, frame = cap.read()
    if not is_read:
        break

    cv2.imwrite(f'{path_to_save}/{str(count).zfill(6)}.png', frame)

    count+=1
    if count%1000 == 0:
        print(count)
```

```python
path_to_save_crop = 'print_attack'
if not os.path.isdir(path_to_save_crop):
        os.mkdir(path_to_save_crop)

count=0
for img in os.listdir(f'/content/{path_to_save}'):

    im = cv2.imread(f'/content/{path_to_save}/{img}')
    im_crop = im[0:720, 330:950]


    cv2.imwrite(f'{path_to_save_crop}/{img}', im_crop)

    count+=1

    if count%1000 == 0:
        print(count)
```
